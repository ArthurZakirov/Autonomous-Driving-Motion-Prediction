```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 7, 0.1)
y = 0.5 * x 
y_exp = (1.13 ** (x*2 + 1)-0.9) 


plt.figure()
plt.grid()
plt.plot(x,y, label='(original)')
plt.plot(x,y_exp, label='(unser) iterative Refinement')
plt.xlim([0, 6]);
plt.ylim([0, 6])
plt.xlabel('time [s]')
plt.ylabel('distance [m]')
plt.legend(loc='best')
```
